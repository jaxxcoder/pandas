```python
import pandas as pd
```

*Elements of our DataFrame

```python
DataFrame_elements={"Model":['hp15','hp14','hp15s'],
                    "Processor":['Ryzen5','Ryzen5','Ryzen7'],
                    "Price":[58000,48000,65000]
    
}
```

## Creating our DataFrame

```python
Data_Frame=pd.DataFrame(DataFrame_elements)
```

```python
Data_Frame
```

## Converting our DataFrame to .CSV file

```python
Data_Frame.to_csv('your_file_name.csv')
```

## Converting out DataFrame to .CSV file without index value

```python
Data_Frame.to_csv('your_file_name.csv',index=False)
```

*Now you can easily read or open this .csv file using microsoft excel or any .csv format supported applications
